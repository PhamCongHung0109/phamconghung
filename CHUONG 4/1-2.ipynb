```python
import numpy as np
#Yêu cẩu 1 :
array_rd = np.random.randint(low=0,high=100,size=(12,12),dtype=int)
print(array_rd)
print("Kiểu dữ liệu của phần tử : ",array_rd.dtype)
print("Kích thước của mảng ma trận : ",array_rd.shape)
print("Số phần tử củ mảng ma trận :",array_rd.size)
print("Số chiều của mảng ma trận :",array_rd.ndim)
#Yêu cầu 2 :
vector_c= np.diag(array_rd)
vector_p = np.diag(np.fliplr(array_rd))
print("Đường chéo chính là :",vector_c)
print("Đường chéo phụ là :",vector_p)
#Yêu cầu 3 :
so=int(input("Nhập vào số nguyên 0-100 :"))
array_bang = np.sum(array_rd == so)
array_nho =np.sum(array_rd < so)
array_lon = np.sum(array_rd > so)
print("Số phần tử có giá trị bằng ",so,"trong ma trận trên là :",array_bang)
print("Số phần tử có giá trị nhỏ hơn ",so,"trong ma trận trên là :",array_nho)
print("Số phần tử có giá trị lớn hơn ",so,"trong ma trận trên là :",array_lon)
                 
```
