## Thực Hành 1 :


```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv("C:/Users/Admin/Downloads/gas_prices.csv")
x = data["Year"]
y = data["Japan"]
# Vẽ biểu đồ
plt.figure(figsize=(10, 5))
plt.plot(x,y, color='teal', marker='D', linestyle='-.', linewidth=2)

# Thiết lập tiêu đề và nhãn trục
plt.title("Biểu đồ giá Gas tại Nhật giai đoạn 1990-2008", fontsize=14)
plt.xlabel("Năm", fontsize=12)
plt.ylabel("Giá Gas (USD)", fontsize=12)

# Giới hạn trục y
plt.ylim(2.5, 6.0)

# Lưới ngang
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Xoay nhãn trục x
plt.xticks(rotation=45)

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```

## Thực Hành 2 :

```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc dữ liệu
data = pd.read_csv("C:/Users/Admin/Downloads/gas_prices.csv")

# Chọn 4 quốc gia cần hiển thị
years = data["Year"]
australia = data["Australia"]
south_korea = data["South Korea"]
uk = data["UK"]
usa = data["USA"]

# Tạo khung hình
plt.figure(figsize=(10, 6))

# Vẽ đường cho từng quốc gia
plt.plot(years, australia, color='deepskyblue', linestyle='--', linewidth=2, label='Australia')
plt.plot(years, south_korea, color='gold', linestyle='--', linewidth=2, label='South Korea')
plt.plot(years, uk, color='red', marker='^', linewidth=2.5, label='UK')        # Giá cao nhất
plt.plot(years, usa, color='black', marker='v', linewidth=2.5, label='USA')   # Giá thấp nhất

# Tiêu đề và nhãn
plt.title("Gas Prices over Time 1990-2008", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Dollars ($)", fontsize=12, fontweight='bold')

# Lưới dọc
plt.grid(axis='x', linestyle='--', linewidth=0.7, alpha=0.7)

# Xoay trục x
plt.xticks(rotation=45)

# Giới hạn trục y
plt.ylim(0, 8)

# Hiển thị chú giải (legend)
plt.legend(loc='lower right', fontsize=10, frameon=True)

# Căn chỉnh bố cục
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

## Thực Hành 3 :

```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
labels = ['K60', 'K61', 'K62', 'K63', 'K64', 'K65']
boys = np.array([200, 340, 260, 440, 300, 180])
girls = np.array([30, 60, 90, 160, 180, 220])
total = boys + girls

# Tổng sinh viên nam và nữ
sex = ['Nam', 'Nữ']
sum_boy = boys.sum()
sum_girl = girls.sum()

# -------- VẼ BIỂU ĐỒ --------
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Biểu đồ tròn
colors = ['#1f77b4', '#ff7f0e']  # Xanh và cam
ax[0].pie(
    [sum_boy, sum_girl],
    labels=sex,
    colors=colors,
    autopct='%1.2f%%',
    startangle=90,
    textprops={'fontsize': 12, 'weight': 'bold'}
)
ax[0].set_title("Biểu đồ tỷ lệ SV Nam - Nữ của Khoa CNTT", fontsize=13)

# Biểu đồ cột
bars = ax[1].bar(labels, total, color=['#1f77b4' if t != max(total) else '#ff7f0e' for t in total])
ax[1].set_title("Biểu đồ số lượng SV theo từng khoá", fontsize=13)
ax[1].set_xlabel("Khoá", fontsize=12)
ax[1].set_ylabel("Số lượng SV", fontsize=12)
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

# Hiển thị giá trị trên cột
for bar in bars:
    height = bar.get_height()
    ax[1].text(bar.get_x() + bar.get_width()/2, height + 10, str(height),
               ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

## Thực Hành 4 :

```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu
labels = ['K60', 'K61', 'K62', 'K63', 'K64', 'K65']
boys = np.array([200, 340, 260, 440, 300, 180])
girls = np.array([30, 60, 90, 160, 180, 220])
total = boys + girls

# ----------- BIỂU ĐỒ ĐƯỜNG -----------
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(labels, total, color='red', marker='s', linestyle='-', linewidth=2, label='Tổng số SV')
plt.plot(labels, boys, color='blue', marker='o', linestyle='--', linewidth=2, label='Sinh viên Nam')
plt.plot(labels, girls, color='black', marker='o', linestyle=':', linewidth=2, label='Sinh viên Nữ')

plt.title("Biểu đồ số lượng SV Nam, Nữ và tổng theo từng Khoá", fontsize=12)
plt.xlabel("Khoá", fontsize=11)
plt.ylabel("Số lượng SV", fontsize=11)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# ----------- BIỂU ĐỒ TRÒN (DONUT) -----------
plt.subplot(1, 2, 2)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
explode = [0.05, 0.05, 0.05, 0.05, 0.05, 0.05]

plt.pie(total, labels=labels, autopct='%1.2f%%', startangle=90,
        colors=colors, wedgeprops={'width': 0.4, 'edgecolor': 'white'},
        explode=explode, textprops={'fontsize': 11, 'weight': 'bold'})

plt.title("Biểu đồ tỷ lệ SV của từng Khoá", fontsize=12)

plt.tight_layout()
plt.show()
```
