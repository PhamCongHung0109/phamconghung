## Thực Hành 3 :

```python
import pandas as pd
import numpy as np
path ="D:/Data_Patient.csv"
df_patient = pd.read_csv(path,index_col = 0)
#Yêu Cầu 3.1 :
print("-------------Thông tin tổng quan-------------")
print(df_patient.info)
print("-------------Thông tin 10 bệnh nhân đầu tiên :")
print(df_patient.head(10))
print("-------------Thông tin 5 bệnh nhận cuối cùng :")
print(df_patient.tail(5))
df_patient.rename(columns={'feature_1':'Age',
                           'feature_2':'Gender',
                           'feature_3':'Type',
                           'feature_4':'Blood_pressure',
                           'feature_5':'Cholesterol',
                           'feature_6':'Heartbeat',
                           'feature_7':'Thalassemia',
                           'feature_8':'Result',},inplace=True)
print("-------------Sau khi đổi tên là :")
print(df_patient.head())
#Yêu Cầu 3.2:
print("Bệnh nhân có tuổi trẻ nhất là :",df_patient['Age'].min())
print("Bệnh nhân có tuổi già nhất là :",df_patient['Age'].max())
print("Cholesterol trung bình là :",df_patient['Cholesterol'].mean())
print("Cholesterol độ lệch chuẩn là :",df_patient['Cholesterol'].std())
print("Có số bệnh nhân nam là :",df_patient['Gender'].value_counts().get('Male'))
print(df_patient.describe())
#Yêu Cầu 3.3:
print("------------Cột dữ liệu chứa missing data:")
print(df_patient.isnull().sum())
#Yều Cầu 3.4:
print("------------Hiển thị 3 bệnh nhận là :")
print(df_patient.loc[['Patient_100','Patient_150','Patient_200']])
#Yêu Cầu 3.5:
df_patient.replace({'Male':'0','Female':'1',0:'No',1:'Yes'},inplace=True)
df_patient.iloc[5,7] =4.0  
print(df_patient.head())
```

## Thực Hành 4 :

```python

#Yêu cầu 4.1:
import matplotlib.pyplot as plt

df_patient.rename(columns={
    'feature_1': 'Age',
    'feature_2': 'Gender',
    'feature_3': 'Type',
    'feature_4': 'Blood_pressure',
    'feature_5': 'Cholesterol',
    'feature_6': 'Heartbeat',
    'feature_7': 'Thalassemia',
    'feature_8': 'Result'
}, inplace=True)

print("Dữ liệu sau khi đổi tên cột:")
display(df_patient.head())
#Yêu cầu 4.2:


df_male = df_patient[df_patient['Gender'] == 'Male']
print("Danh sách bệnh nhân Nam:")
display(df_male.head())

df_female = df_patient[df_patient['Gender'] == 'Female']
print("Danh sách bệnh nhân Nữ:")
display(df_female.head())

df_no = df_patient[df_patient['Result'] == 0]
print("Danh sách bệnh nhân KHÔNG bị đau tim:")
display(df_no.head())

df_yes = df_patient[df_patient['Result'] == 1]
print("Danh sách bệnh nhân CÓ bị đau tim:")
display(df_yes.head())

# Yêu cầu 4.3:


df_cond1 = df_patient[(df_patient['Result'] == 1) & (df_patient['Age'] > 70)]
print("Bệnh nhân mắc bệnh đau tim và trên 70 tuổi:")
display(df_cond1)

df_cond2 = df_patient[(df_patient['Gender'] == 'Female') & 
                      (df_patient['Blood_pressure'] > 170) & 
                      (df_patient['Result'] == 0)]
print("Bệnh nhân Nữ, huyết áp > 170, không bị đau tim:")
display(df_cond2)

df_cond3 = df_patient[(df_patient['Type'] == 'Typical angina') & 
                      (df_patient['Gender'] == 'Male') & 
                      (df_patient['Result'] == 1)]
print("Bệnh nhân Nam, Typical angina và bị đau tim:")
display(df_cond3)
#Yêu cầu 4.4: Xác định:

print("Thống kê huyết áp (Blood_pressure):")
print("Min:", df_patient['Blood_pressure'].min())
print("Max:", df_patient['Blood_pressure'].max())
print("Mean:", df_patient['Blood_pressure'].mean())
print("Median:", df_patient['Blood_pressure'].median())
print("Std:", df_patient['Blood_pressure'].std())

print("\nThống kê nhịp tim (Heartbeat):")
print("Min:", df_patient['Heartbeat'].min())
print("Max:", df_patient['Heartbeat'].max())
print("Mean:", df_patient['Heartbeat'].mean())
print("Median:", df_patient['Heartbeat'].median())
print("Std:", df_patient['Heartbeat'].std())

#Yêu cầu 4.5: 

print("\nSố giá trị khác nhau của Type:", df_patient['Type'].nunique())
print("Giá trị và tần suất:")
display(df_patient['Type'].value_counts())

plt.figure(figsize=(8,5))
df_patient['Type'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Thống kê số lượng theo từng giá trị của Type")
plt.xlabel("Type")
plt.ylabel("Số lượng")
plt.xticks(rotation=30)
plt.show()
```
