```python
N = int(input("Nhập vào một số tự nhiên N (>0): "))
danh_sach_ket_qua = N
danh_sach_nhi_phan = []

if N == 0:
    danh_sach_nhi_phan.append(0)
else:
    temp = N
    while temp > 0:
        du = temp % 2
        danh_sach_nhi_phan.append(du)
        temp = temp // 2

danh_sach_nhi_phan.reverse()

danh_sach_chuyen = ""
for bit in danh_sach_nhi_phan:
    danh_sach_chuyen += str(bit)

print(f"{danh_sach_ket_qua} (hệ 10) = {danh_sach_chuyen} (hệ 2)")
```

```python
def greeting(ho_ten, nam_sinh):
    nam_hien_tai = 2025 
    tuoi = nam_hien_tai - nam_sinh
    return f"Chào bạn {ho_ten}! Năm nay bạn {tuoi} tuổi."
print ("1.Greeting: " + greeting("Nguyễn Văn A",2005))
def rabbit_count(month):
    if month <= 0:
        return 0
    if month == 1 or month == 2:
        return 1
    
    a, b = 1, 1
    for i in range(3, month + 1):
        a, b = b, a + b
    return b

print("2. Rabbit count tháng 6 =", rabbit_count(6))
def count_mark(list_mark):
    hoc_lai = 0
    for mark in list_mark:
        if mark < 5:
            hoc_lai += 1
    tong_sv = len(list_mark)
    return hoc_lai, tong_sv

print("3. Kết quả:", count_mark([9, 4, 6, 3, 8, 10]))
def bmi_show(chieu_cao, can_nang):
    bmi = can_nang / (chieu_cao**2)
    
    if (bmi <= 18.5):
        msg = "Underwhight"
    elif (18.5 <bmi<= 24.9):
        msg = "Normal Weigh"
    elif bmi (24.9 <bmi<= 29.9):
        msg = "Overweight"
    else:
        msg = "Obese"
    
    return f"BMI = {bmi:.2f} → {msg}"

print("4. BMI:", bmi_show(1.7, 60))
def cal_point(list_point):
    tb10 = sum(list_point) / len(list_point)
    tb4 = tb10 / 2.5
    return round(tb10, 2), round(tb4, 2)

print("5. Điểm trung bình:", cal_point([8, 7, 9, 6]))
def list_prime(n):
    if n < 2:
        return []
    prime = [True] * (n + 1)
    prime[0] = prime[1] = False
    p = 2
    while p * p <= n:
        if prime[p]:
            for i in range(p*p, n+1, p):
                prime[i] = False
        p += 1
    return [i for i in range(n+1) if prime[i]]
print("6. Danh sách số nguyên tố:", list_prime(30))
```

```python
class Person:
    def __init__(self, name, year, height, weight):
        self.name = name
        self.year = year
        self.height = height
        self.weight = weight

    def Greeting(self):
        print("My name is", self.name)
        print("I am", 2024 - self.year, "years old. Nice to meet you!")

    def BMI(self):
        return round(self.weight / (self.height ** 2), 2)


# Demo ý 1
p1 = Person("Đặng Văn Nam", 1985, 1.65, 59)

p1.Greeting()
print("BMI:", p1.BMI())
class Person:
    def __init__(self, name, year, height, weight):
        self.name = name
        self.year = year
        self.height = height
        self.weight = weight

    def Greeting(self):
        print("My name is", self.name)
        print("I am", 2024 - self.year, "years old.")

    def BMI(self):
        return round(self.weight / (self.height**2), 2)


class Vietnam(Person):
    def __init__(self, name, year, height, weight, nation="Kinh"):
        super().__init__(name, year, height, weight)
        self.nation = nation

    def Greeting(self):
        print("Tôi tên là:", self.name, "- Dân tộc:", self.nation)
        print("Tôi sinh năm", self.year, ". Rất vui được gặp bạn!")


vn1 = Vietnam("Thanh Huyền", 2000, 1.69, 52, "Tày")

vn1.Greeting()
print("BMI:", vn1.BMI())
```
